```python
from plotly import express as px
import pandas as pd
from tqdm.auto import tqdm, trange
from nabeatsu import check_nabeatsu
tqdm.pandas()
```

# check nabeatsu

```python
df = pd.DataFrame(range(10000))
df.columns = ["value"]
df["nabeatsu"] = df["value"].progress_apply(check_nabeatsu)
df
```

```python
df["nabeatsu"].value_counts(normalize=True)
```

# cumsum

```python
def count_num(th, df=df):
    tmp = df.iloc[:th]
    tmp_vc = pd.Series(tmp["nabeatsu"].value_counts(normalize=True).to_dict())
    tmp_vc.loc["th"] = th
    return pd.DataFrame(tmp_vc).T
```

```python
vc = pd.Series(range(len(df))).progress_apply(count_num)
vc = pd.concat(list(vc))
vc
```

```python
px.line(
    vc,
    x="th",
    y=[True, False]
)
```
